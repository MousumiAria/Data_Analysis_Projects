### Feature engineering
#### 1. Here I am going to create one new column called Family_size by adding SibSp and Parch column

```python
df['Family_Size']=df['SibSp']+df['Parch']

def Family_Type(i):
    if i == 0:
        return 'Alone'
    elif i>0 and i<=4:
        return 'Middle'
    else:
        return 'Large'
```

```python
df['Family_Type']=df['Family_Size'].apply(Family_Type)
```

```python
df=df.drop(columns=['SibSp','Parch','Family_Size'])
```

### Dectecting Outliers
#### For Numerical columns
##### 1. If the data is followong normal distribution, any data beyond 3SD(Standered Deviation) - mean + 3SD  can be considered as outlier.
##### 2. If the data dose not following normal distribution, using boxplot we can eleminate points beyond Q1 - 1.5 IQR and Q3 + 1.5 IQR
#### For Categorical Column
##### If the data is highly imbalance then we can remove the outliers 

```python
# Handling Outlier in Age column
df=df[df['Age']<(df['Age'].mean()+ 3*df['Age'].std())]
df.shape
```

```python
# Handling Outlier from Fare column
# Using IQR method

Q1=np.percentile(df['Fare'],25)
Q3=np.percentile(df['Fare'],75)

Outlier_low=Q1-1.5*(Q3-Q1)
Outlier_high=Q3+1.5*(Q3-Q1)

df=df[(df['Fare']>Outlier_low) & (df['Fare']<Outlier_high)]
```

#### One Hot Encoding
##### Transform catagorical values into numbers for columns Pclass, Sex, Embarked and Family_Type  

```python
df.sample(4)
```

```python
df=pd.get_dummies(data=df, columns=['Pclass','Sex', 'Embarked', 'Family_Type'], drop_first=True)
#df
```
